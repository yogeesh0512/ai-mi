```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv('C:/Users/GPT BANTWAL/Documents/apples_and_oranges.csv')
data
```

```python
import seaborn as sns
sns.scatterplot(x="Weight",y="Size",hue="Class",data=data)
```

```python
#splitting data into training and test set
training_set,test_set=train_test_split(data,test_size=0.2,random_state=1)
```

```python
#prepare data for applying it to svm
x_train=training_set.iloc[:,0:2].values
y_train=training_set.iloc[:,2].values
x_test=test_set.iloc[:,0:2].values
y_test=test_set.iloc[:,2].values
```

```python
#fitting the data(train a model)
classifier=SVC(kernel='rbf',random_state=1,C=1,gamma='auto')
classifier.fit(x_train,y_train)
```

```python
#Perform prediction on x_test data
y_pred=classifier.predict(x_test)
print(y_pred)
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy=float(cm.diagonal().sum())/len(y_test)
print('model accuracy is:',accuracy*100,'%')
```
