```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
credit_data=pd.read_csv('C:/Users/GPT BANTWAL/Documents/credit_risk.csv')
credit_data
```

```python
credit_data.info()
```

```python
x=credit_data.columns.drop("class")
x
```

```python
y=credit_data['class']
y
```

```python
credit_data_encoded=pd.get_dummies(credit_data[x])
print("Total number of predictors after encoding =",len(credit_data_encoded.columns))
credit_data_encoded.columns
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(credit_data_encoded,y,test_size=0.15,random_state=100)
```

```python
print("xtrain:",xtrain.shape)
print("xtest:",xtest.shape)
print("ytrain:",ytrain.shape)
print("ytest:",ytest.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(random_state=1)
model.fit(xtrain,ytrain)
```

```python
train_predictions=model.predict(xtrain)
test_predictions=model.predict(xtest)
```

```python
from sklearn.tree import export_graphviz
import graphviz
```

```python
credit_data=export_graphviz(model,out_file=None,feature_names=credit_data_encoded.columns,class_names=model.classes_,)
graph=graphviz.Source(credit_data)
graph
```

```python
train_accuracy=model.score(xtrain,ytrain)
print("Accuracy of the model on train data=",train_accuracy)
test_accuracy=model.score(xtest,ytest)
print("Accuracy of the model on train data=",test_accuracy)
```

```python
model1=DecisionTreeClassifier(min_samples_split=10,min_impurity_decrease=0.005)
model1.fit(xtrain,ytrain)
print("train_accuracy=",model1.score(xtrain,ytrain))
print("test_accuracy=",model1.score(xtest,ytest))
```
